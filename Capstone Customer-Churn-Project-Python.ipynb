# 3. Linear Regression


```python
# Build a simple model where dependent variable is ‘MonthlyCharges’
# and independent variable is ‘tenure’:
# a. Divide the dataset into train and test sets in 70:30 ratio.
# b. Build the model on train set and predict the values on test set
# c. After predicting the values, find the root mean square error
# d. Find out the error in prediction & store the result in ‘error’
# e. Find the root mean square error
```

## Data Preprocessing

```python
df.head()
```

```python
#label encoding for labelling all categorical variable with a fixed value
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in df.columns :
    if df[i].dtype == 'object' :
        df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

## Feature Engineering

```python
X=pd.DataFrame(df['tenure']) #feature
y=pd.DataFrame(df['MonthlyCharges']) #target
```

```python
X
```

```python
y
```

### Train and test the Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30, random_state=52)
```

### Model Building

```python
from sklearn.linear_model import LinearRegression
linreg= LinearRegression()
```

```python
linreg.fit(X_train,y_train)
```

```python
#model prediction
```

```python
y_pred=linreg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error,mean_absolute_error
rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse
```

```python
mse= mean_squared_error(y_test, y_pred)
mse
```

```python
mape = mean_absolute_percentage_error(y_test, y_pred)
mape
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Absolute % Error:', mean_absolute_percentage_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
error_pred = pd.DataFrame(columns=['Actual_data', 'Prediction_data'])
error_pred['Actual_data']=y_test
error_pred['Prediction_data']=y_pred
error_pred['Error']=error_pred['Actual_data']-error_pred['Prediction_data']
error_pred
```

```python
plt.figure(figsize=(5,5))
plt.scatter(error_pred['Actual_data'], error_pred['Prediction_data'], c='purple')
```

```python
#goodness of fit of a regression model
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
